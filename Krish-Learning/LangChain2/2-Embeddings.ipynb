
# Faiss - Facebook AI Similarity Search - It's a library for efficient similarity search and clustering of dense vectors.
 It contains algorithms that search in sets of vectors of any size, up to ones that possibly do not fit in RAM.
 It also contains supporting code for evaluation and parameter tuning. 


```python
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter 
from langchain_community.vectorstores import FAISS

## Load the text into documents. 
loader = TextLoader('speech.txt')
documents = loader.load()

## Load documents and split into chunks. 
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_documents(documents) 

## Load document chunks into vector store. 
embeddings = OllamaEmbeddings(model="gemma:2b")
db = FAISS.from_documents(docs, embeddings)

docs2 = db.similarity_search("How does you feel about coming back to school?")

## -- Retrievals. 

retrieval = db.as_retriever()
docs3 = retrieval.invoke("How does you feel about coming back to school?")

docs_and_scrore = db.similarity_search_with_score("How does you feel about coming back to school?")
docs_and_scrore

```

```python
###  -- Saving and loading embeddings 
db.save_local("faiss_index")

new_db = FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)
docs = new_db.similarity_search("How does you feel about coming back to school?") 
docs
```

# Chroma - 
Chroma is an AI-native Open source vector database focused on developer productivity. 



```python
from langchain_chroma import Chroma

from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter 


## Load the text into documents. 
loader = TextLoader('speech.txt')
documents = loader.load()

## Load documents and split into chunks. 
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_documents(documents) 

## Load document chunks into vector store. 
embeddings = OllamaEmbeddings(model="gemma:2b")
vectordb = Chroma.from_documents(documents=docs, embedding=embeddings, persist_directory="./chroma_db")
vectordb



```

```python
## Query it. 

query = "you are now part of a community"
query_result2 = vectordb.similarity_search(query)
query_result2
```
